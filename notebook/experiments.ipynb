```python
documents = SimpleDirectoryReader("../Data")
```

```python
doc = documents.load_data()
```

```python
print(doc[0].text)
```

```python
model = Gemini(models = 'gemini-1.5-flash', api_key=google_api_key)
```

```python
gemini_embed_model = GeminiEmbedding(model = "models/embedding-004")
```

```python
from llama_index.core import Settings
from llama_index.core import VectorStoreIndex

Settings.llm = model
Settings.embed_model = gemini_embed_model
Settings.chunk_size = 800
Settings.chunk_overlap = 20

# Then use Settings directly or create your index
index = VectorStoreIndex.from_documents(doc)
```

```python
index.storage_context.persist()
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("Thera ere how many people who bought the land and which year and how much quantity")
```

```python
print(response.response)
```
